**Extract Name, rating, reviews, public/pivate, headquarter, how old,no of employee, tags, about from ambitionbox website**

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
company = soup.find_all('div', class_='company-content-wrapper')
```

```python
len(company)
```

```python
name = []
rating = []
review = []
ctype = [] #Company type
hq = []
old =[]
employee =[]

for i in company:
  #print(i.find('h2').text.strip())   # This will print the name of all the companies
   name.append(i.find('h2').text.strip())
   rating.append(i.find('p', class_='rating').text.strip())
   review.append(i.find('a',class_='review-count').text.strip())
   ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
   hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
   old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
   employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())

```

```python
print("Names of companies are: ", name)
print("Ratings of companies are: ", rating)
print("Reviews of companies are: ", review)
print("company types are: ", ctype)
print("head quarters of the comapnies are : ",hq)
print("Age of the companies are: ", old)
print("No of employee in the companies are: ", employee)
```

```python
# Creating DataFrame now using all the data
d = {'name':name, 'rating':rating, 'review':review, 'type':ctype, 'hq':hq, 'age':old, 'No. of employee':employee}
df = pd.DataFrame(d)
df.head()
```

```python
# Creating dataframe for all the pages

final = pd.DataFrame()

for j in range(1,11):
  url = "https://www.ambitionbox.com/list-of-companies?page={}".format(j)
  webpage = requests.get(url).text

  soup = BeautifulSoup(webpage,'lxml')

  company = soup.find_all('div', class_='company-content-wrapper')

  name = []
  rating = []
  review = []
  ctype = [] #Company type
  hq = []
  old =[]
  employee =[]

  for i in company:
    #print(i.find('h2').text.strip())   # This will print the name of all the companies
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    review.append(i.find('a',class_='review-count').text.strip())
    try:
      ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:
      hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    try:
      old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      old.append(np.nan) 
    try:
      employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      employee.append(np.nan)

    # Creating DataFrame now using all the data
    d = {'name':name, 'rating':rating, 'review':review, 'type':ctype, 'hq':hq, 'age':old, 'No. of employee':employee}
    df = pd.DataFrame(d)
    final = final.append(df, ignore_index = True)  # Ignore index is set as False so that after fetching 10 records from page 1 when it fetches from page 2 the index in continued as 11 and not again starts from 1

```
