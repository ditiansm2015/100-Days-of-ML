```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

import pandas as pd
import numpy as np
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
```

```python
# We pass columns inside a list of tupples

transformer = ColumnTransformer(transformers = [
    ('tnf1', SimpleImputer(), ['fever']),   # Applying SimpleImputer in 'fever' column to replace missing values with mean
    ('tnf2', OrdinalEncoder(categories = [['Mild','Strong']]),['cough']),   # applying Oridinal encoding on 'cough' column
    ('tnf3', OneHotEncoder(sparse=False, drop ='first'), ['gender','city']) # applying OHE on 'gender' and 'city' column
], remainder = 'passthrough')   # Remainder = passthrough means all other columns should be as it is while remainder = 'drop' means all the other columns would be dropped
```

```python
# Applying transformation/encoding on X_train data

transformer.fit_transform(X_train)
```
