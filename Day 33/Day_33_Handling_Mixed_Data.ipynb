**Mixed Data** - A column having both categorical as well as numerical data
Eg - Train compartment number - B5 may mean B class and 5th compartment

### **How to deal with mixed data?**
Create two columns from one mixed data column:

*   In one column keep categorical data (extract from mix data) and put may  
    be -9999 in its corresponding Numerical colum
*   In second colum keep numerical data (extract from mix data) and may be 
    put "missing" in categorical columm





**We will work on titanic dataset - number column, cabin column and ticket column to show how to handle mix variable dataset**


Note we have done some modifications in original titanic dataset to create mixed dataset (column - number).

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title("Passengers travelling with")
```

```python
# Working on number column

# Extract numerical part from number column
df['number_numerical'] = pd.to_numeric(df['number'], errors = "coerce", downcast = 'integer')
```

```python
# Extract categorical part from number column
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)  # Wherever there is a null value in number_numerical column, replace the null with nan

df.head()
```

```python
# Working on Cabin column

df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')  # the reg expression '\d+' extracts the numeric part of data
df['cabin_cat'] = df['Cabin'].str[0]  # captures the first letter
```

```python
df.head()
```

```python
# Working on ticket column

# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
